## 設定模型的存放路徑, 讀取先前的訓練結果到model

```python
#set model_path
model_path = "/content/drive/My Drive/Hands on Deep Learning/mask_dataset_preprocessing/model-logs/basic_model-last-model.h5"
```

```python
#load model
model = keras.models.load_model(model_path)
```

```python
model.summary()
```

## 由於模型已經訓練完成, 因此我們只匯出模型的架構跟權重以減少檔案大小, 方便上傳到Github上

```python
#save model with structure and weights
model.save_weights("/content/drive/My Drive/Hands on Deep Learning/mask_dataset_preprocessing/model-logs/best_model.h5")#save model with structure and weights
json_string = model.to_json()

with open("/content/drive/My Drive/Hands on Deep Learning/mask_dataset_preprocessing/model-logs/model.config", "w") as text_file:    
    text_file.write(json_string)
```

```python
#把剛剛儲存的參數讀到新的model測試看看
with open("/content/drive/My Drive/Hands on Deep Learning/mask_dataset_preprocessing/model-logs/model.config", "r") as text_file:
    json_string = text_file.read()
model2 = keras.models.model_from_json(json_string)
model2.load_weights("/content/drive/My Drive/Hands on Deep Learning/mask_dataset_preprocessing/model-logs/best_model.h5")
```
