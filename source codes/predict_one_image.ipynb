```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input 
import numpy as np
```

## Case 1. 如果是透過model.save輸出.h5的模型結果

### 設定影像跟模型路徑

```python
img_size = 224
# img_path = "1blue_penguin.jpg"
# img_path = "1penguin.jpg"
# img_path = "king_penguin_6.jpg"
img_path = "DDG_3.jpg"
model_path = 'best_model.h5'
```

### 讀取模型 -> 資料前處理 -> 預測模型

```python
# load model
model = keras.models.load_model(model_path)

# preprocess image
image = tf.keras.preprocessing.image.load_img( img_path, target_size=(img_size, img_size))
input_arr = keras.preprocessing.image.img_to_array(image)
input_arr = preprocess_input( np.array([input_arr]) ) # Convert single image to a batch.

# predict on image
predictions = model.predict(input_arr)
print( predictions, predictions.argmax() )
```

## Case 2. 如果模型已經輸出成架構跟權重

```python
img_size = 224
# img_path = "1blue_penguin.jpg"
# img_path = "1penguin.jpg"
# img_path = "king_penguin_6.jpg"
img_path = "DDG_3.jpg"
model_config_path = './models/model.config'
model_weight_path = './models/best_model.h5'
```

```python
# load model
with open(model_config_path, "r") as text_file:
    json_string = text_file.read()
model = keras.models.model_from_json(json_string)
model.load_weights(model_weight_path)

# preprocess image
image = tf.keras.preprocessing.image.load_img( img_path, target_size=(img_size, img_size))
input_arr = keras.preprocessing.image.img_to_array(image)
input_arr = preprocess_input( np.array([input_arr]) ) # Convert single image to a batch.

# predict on image
predictions = model.predict(input_arr)
print( predictions, predictions.argmax() )
```
