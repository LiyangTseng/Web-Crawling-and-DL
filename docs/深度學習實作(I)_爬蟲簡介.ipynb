#### 範例：辦公室的下午茶 (資料來源https://pala.tw/python-web-crawler/)
請想像你身為辦公室的手搖杯飲料訂購對外窗口。

下午三點半，眼前正擺著剛送來的20杯手搖飲料，因為劇情需要，你想著這20杯飲料，不就像網頁上的20筆資料嗎？而每杯飲料上的標籤，例如去冰、半糖、微糖，正如資料上所貼的分類標籤class。

但因為字實在太小了，為了節省大家拿取的時間，你決定用筆在飲料上寫下每個人的名字id，啊，這杯是老闆的、這杯是Pala的…。

好的，從以上的例子我們可以知道分類標籤class是可以重複出現的，就像很多杯飲料都會出現半糖；而id是唯一的，可以是名字、或是其他方便識別的內容。



#### 實作練習: 觀察定位元素

請在網頁https://pala.tw/class-id-example/ 上點右鍵，選擇檢查，點觀察器後指定元素：

選擇class，要在開頭加上句點.，例如.半糖；選擇id時，則要在前面加上井字號#，#老闆、#Pala，直接複製貼上這個程式碼試試看：

```python
import requests
from bs4 import BeautifulSoup

tag = input("請輸入定位元素，class前面加上.，id前面加上# ")
res = requests.get('http://pala.tw/class-id-example/')
soup = BeautifulSoup(res.text, "lxml")

for drink in soup.select('{}'.format(tag)):
    print(drink.get_text())
```

### 範例: PTT爬蟲

```python
import requests
from bs4 import BeautifulSoup
```

```python
#設定看板名稱
board_name = "Baseball"
```

```python
url = "https://www.ptt.cc/bbs/" + board_name + "/index"
print(url)
```

### 先找出規律性

目標網址看起來像是 https://www.ptt.cc/bbs/Baseball/index[數字].html/ 

```python
#測試看看
page_num = 10000
```

```python
res = requests.get(url + str(page_num) + ".html") #發送要求
print(res) #看看收到的response是不是200
```

```python
res_text = res.text #把收到的資訊取出來, 存到res_text這個變數
print(res_text) #印出來看看
```

```python
res_soup = BeautifulSoup(res_text, "html.parser") #使用beautifulsoup來整理收到的資訊
print(res_soup) #印出來看看
```

```python
#把每個title整理出來 就是這一頁的所有文章標題了!!
for label in res_soup.findAll("div", {"class": "title"}):
    label_txt = label.text
    label_txt = str(label_txt).replace("\n", "") #把換行符號取代掉
    print(label_txt)
```

### 測試完畢, 這時候來自動爬五頁文章看看

```python
#設定從哪頁開始, 要爬幾頁
page_start = 10010
page_num = 5

for i in range(page_num):
    res = requests.get(url + str(page_start) + ".html")
    res_text = res.text
    res_soup = BeautifulSoup(res_text, "html.parser")
    
    for label in res_soup.findAll("div", {"class": "title"}):
        label_txt = label.text
        label_txt = str(label_txt).replace("\n", "")
        print(label_txt)
    page_start -= 1 
```

### 從網路上抓取圖片

```python
from selenium import webdriver
import time
import urllib
import os

# 存圖位置
local_path = '/home/daniel/HandsOnDL/images/wearing_masks'

# 爬取頁面網址 
url = 'https://www.google.com/search?q=wearing+mask+&tbm=isch&ved=2ahUKEwjQoorM3pzrAhWjI6YKHZIgAAIQ2-cCegQIABAA&oq=wearing+mask+&gs_lcp=CgNpbWcQAzIECCMQJzIECCMQJzICCAAyAggAMgIIADICCAAyAggAMgIIADICCAAyAggAOgQIABATOggIABAFEB4QEzoECAAQHlCaBVj1CWDbygFoAHAAeACAATiIAegBkgEBNZgBAKABAaoBC2d3cy13aXotaW1nwAEB&sclient=img&ei=zJU3X9CMKqPHmAWSwYAQ&bih=654&biw=1235'
# 啟動chrome瀏覽器
chrome_path = "/home/daniel/Documents/chromedriver" #chromedriver.exe執行檔所存在的路徑
driver = webdriver.Chrome(chrome_path) 
  
# 最大化窗口，因為每一次爬取只能看到視窗内的圖片  
driver.maximize_window()  

# 紀錄下載過的圖片網址，避免重複下載  
img_url_dic = {}  
  
# 瀏覽器打開爬取頁面
driver.get(url)  

# 模擬滾動視窗瀏覽更多圖片
pos = 0  
m = 0 # 圖片編號 
for i in range(5):  
    pos += i*500 # 每次下滾500  
    js = "document.documentElement.scrollTop=%d" % pos  
    driver.execute_script(js)  
    time.sleep(5)    
    
    for _ in range(10):
        xpath = '/html/body/div[2]/div[2]/ul/li['+str(m+2)+']/div/a[1]/img'
        try:
            for element in driver.find_elements_by_xpath(xpath):
                img_url = element.get_attribute('src')
                print(img_url)
            
                # 保存圖片到指定路徑
                if img_url != None and not img_url in img_url_dic:
                    img_url_dic[img_url] = ''  
                    m += 1
                    # print(img_url)
                    ext = img_url.split('/')[-1]
                    # print(ext)
                    filename = str(m) + 'kerGee' + '_' + ext +'.jpg'
                    print(filename)

                    # 保存圖片
                    urllib.request.urlretrieve(img_url, os.path.join(local_path , filename))                    
        except OSError:
            print('發生OSError!')
            print(pos)
            break;

        time.sleep(3)
        
driver.close()
```
